```python
# df = df[df['severity'] != 'normal']
df = df[df['severity'] != 'enhancement']
# df = df[df['severity'] != 'normal']


```

```python
#0.to lowercase
def convert_description_to_lowercase(df):
    # Create a copy of the DataFrame to avoid modifying the original
    modified_df = df.copy()
    
    # Convert the text in the 'description' column to lowercase
    modified_df['description'] = modified_df['description'].str.lower()
    
    return modified_df
```

```python
grouped = df.groupby('severity')
sampled_df = pd.DataFrame()

for name, group in grouped:
    sampled_group = group.sample(n=round(60000/len(grouped)), random_state=42 , replace=True)
    sampled_df = pd.concat([sampled_df, sampled_group])

sampled_df.reset_index(drop=True, inplace=True)


df_wrn = convert_description_to_lowercase(df)
df_wrn.head(10)
```

```python
#1.removing numbers from the data
import re
import pandas as pd

def remove_numbers_from_description(df):
    # Create a copy of the DataFrame to avoid modifying the original
    modified_df = df.copy()
    
    # Convert non-string values to strings
    modified_df['description'] = modified_df['description'].apply(lambda x: str(x))
    
    # Remove any digits from the description column
    modified_df['description'] = modified_df['description'].apply(lambda x: re.sub(r'\d+', '', x))
    
    return modified_df
```

```python
 
df_wrn = remove_numbers_from_description(df_wrn)
df_wrn.head(10)
```

```python
import re
df_wrn["description"] = df_wrn["description"].astype(str)

# Suppose your DataFrame is named df, and the column containing text is named "description"
df_wrn["description"] = df_wrn["description"].apply(lambda x: re.sub(r"[^a-zA-Z0-9\s]", "", x))

```

```python
from sklearn.preprocessing import LabelEncoder

# Create an instance of LabelEncoder
le = LabelEncoder()

# Fit and transform the "severity" column
df_wrn["severity_n"] = le.fit_transform(df_wrn["severity"])

# Print the unique values of the encoded column
print(df_wrn["severity_n"].unique())


```
